```python
### karakteristike elektromotorja ###
P_b = 4000 #W
n_m = 1450 #/min

### Dimenzije bobna ###
D_b = 0.23 #m # premer bobna
t_b = 8  # mm # debelina stene
```


## Določitev prestavnega razmerja ##
Preračun je namenjen rešitvi **z sestavljenim 1AI gonilom**:

<img src="Slike\Osnovni_koncept_4.png" width="700" align="center"/>

```python
omega_b=2*v_b/D_b
n_b = 60*omega_b/(2*np.pi)
i = n_m/n_b # celotna potrebna prestava
i1s_1 = 4 # prestava prve stopnje
i1s_2 = 3.5 # prestava druge stopnje
i13_3 = -2.5 # prestava druge stopnje
print(f"Celotno prestavno razmerje je i = {np.around(i, 2)}")
print(f"S prvo stopnjo dosežemo i1s_1 = {i1s_1}, z drugo i1s_2 = {i1s_2} in tretjo i13_3 = {i13_3} je celotno prestavno razmerje i = {i1s_1*i1s_2*i13_3} .")
print(f"({-np.around(100-(-i1s_1*i1s_2*i13_3)/i*100, 2)}% napake)")
```

## Število zob zobnikov in planetnikov ##

```python
### Prva zobniška stopnja ###
# št. zob
z_11 = 28 #min. 17, če ne želimo profilnega pomika
z_31 = z_11*(i1s_1-1)
z_21 = (z_31-z_11)/2
n_p1 = 4 # št. planetnikov
print(f"PRVA ZOBNIŠKA STOPNJA (i1s={i1s_1}):")
print(f"z1 = {z_11}")
print(f"z2 = {z_21}")
print(f"z3 = {z_31}")
print(f"Število planetnikov n_p = {n_p1}.")
print(f"Pogoj za simetrijo z1+z3/np = {z_11+z_31/n_p1} = celo")
print()

### Druga zobniška stopnja ###
# št. zob
z_12 = 36
z_32 = z_12*(i1s_2-1)
z_22 = (z_32-z_12)/2
n_p2 = 6 # št. planetnikov
print(f"DRUGA ZOBNIŠKA STOPNJA (i1s={i1s_2}):")
print(f"z1 = {z_12}")
print(f"z2 = {z_22}")
print(f"z3 = {z_32}")
print(f"Število planetnikov n_p = {n_p2}.")
print(f"Pogoj za simetrijo z1+z3/np = {z_12+z_32/n_p2} = celo")
print()

### Tretja zobniška stopnja ###
# št. zob
z_13 = 36
z_33 = -z_13*i13_3
z_23 = (z_33-z_13)/2
n_p3 = 6 # št. planetnikov
print(f"TRETJA ZOBNIŠKA STOPNJA (i1s={i13_3}):")
print(f"z1 = {z_13}")
print(f"z2 = {z_23}")
print(f"z3 = {z_33}")
print(f"Število planetnikov n_p = {n_p3}.")
print(f"Pogoj za simetrijo z1+z3/np = {z_13+z_33/n_p3} = celo")
```

## Vrtilne frekvence planetnega gonila ##

```python
### Prva zobniška stopnja ###
n_11 = n_m # /min
n_31 = 0
n_s1 = n_11/(1+z_31/z_11)
n_21 = (1-z_31/z_21)*n_s1
print("PRVA ZOBNIŠKA STOPNJA:")
print(f"n1 = {np.around(n_11,2)}/min")
print(f"n3 = {np.around(n_31,2)}/min")
print(f"nS = {np.around(n_s1,2)}/min")
print()

### Druga zobniška stopnja ###
n_12 = n_s1 # /min
n_32 = 0
n_s2 = n_12/(1+z_32/z_12)
n_22 = (1-z_32/z_22)*n_s2
print("DRUGA ZOBNIŠKA STOPNJA:")
print(f"n1 = {np.around(n_12,2)}/min")
print(f"n3 = {np.around(n_32,2)}/min")
print(f"nS = {np.around(n_s2,2)}/min")
print()

### Tretja zobniška stopnja ###
n_13 = n_s2 # /min
n_33 = -(z_13/z_33)*n_13
n_s3 = 0
n_23 = (1-z_33/z_23)*n_s3
print("TRETJA ZOBNIŠKA STOPNJA:")
print(f"n1 = {np.around(n_13,2)}/min")
print(f"n3 = {np.around(n_33,2)}/min")
print(f"nS = {np.around(n_s3,2)}/min")
print()

v_test = D_b*np.pi/60*n_33
print(f"Na izbranem premeru bobna bo hitrost {np.around(v_test,5)}m/s, kar je {-np.around(100+v_b/v_test*100, 2)}% napake od zahteve.")
```

## Vrtilni momenti na posameznih gredeh (brez izgub) ##

**Wolfova shema za gonilo 1AI**
<img src="Slike\Wolfova_shema.png" width="300" align="center"/>

Veljati mora: $M_{t1}+M_{t3}+M_{tS}=0$

```python
### Prva zobniška stopnja ###
M_t11 = P_b/(2*np.pi*n_m/60)*1000 # N mm - glede na moment motorja
i13_1 = -z_31/z_11
M_t31 = -i13_1 * M_t11
M_ts1 = -(1-i13_1)*M_t11
print("PRVA ZOBNIŠKA STOPNJA:")
print(f"M_t1 = {np.around(M_t11/1000,2)} Nm")
print(f"M_t3 = {np.around(M_t31/1000,2)} Nm")
print(f"M_tS = {np.around(M_ts1/1000,2)} Nm")
print(f"Pogoj izpolnjen (vsota = 0)? M_t1 + M_t3 + M_tS = {M_t11+M_t31+M_ts1} N mm" )
print()

### Druga zobniška stopnja ###
M_t12 = M_ts1 # N mm - glede na moment motorja
i13_2 = -z_32/z_12
M_t32 = -i13_2 * M_t12
M_ts2 = -(1-i13_2)*M_t12
print("DRUGA ZOBNIŠKA STOPNJA:")
print(f"M_t1 = {np.around(M_t12/1000,2)} Nm")
print(f"M_t3 = {np.around(M_t32/1000,2)} Nm")
print(f"M_tS = {np.around(M_ts2/1000,2)} Nm")
print(f"Pogoj izpolnjen (vsota = 0)? M_t1 + M_t3 + M_tS = {M_t12+M_t32+M_ts2} N mm" )
print()

### Tretja zobniška stopnja ###
M_t13 = M_ts2 # N mm - glede na moment motorja
i13_3 = -z_33/z_13
M_t33 = -i13_3 * M_t13
M_ts3 = -(1-i13_3)*M_t13
print("TRETJA ZOBNIŠKA STOPNJA:")
print(f"M_t1 = {np.around(M_t13/1000,2)} Nm")
print(f"M_t3 = {np.around(M_t33/1000,2)} Nm")
print(f"M_tS = {np.around(M_ts3/1000,2)} Nm")
print(f"Pogoj izpolnjen (vsota = 0)? M_t1 + M_t3 + M_tS = {M_t13+M_t33+M_ts3} N mm" )

F_test = np.abs(M_t33)/(D_b*1000/2)*ni_pog
print()
print(f"Na izbranem premeru bobna bo sila (z upoštevanjem izgube) {np.around(F_test,2)}N, kar je {np.around(100-F_b/(F_test)*100, 2)}% napake od zahteve.")
```

### preliminarni preračun za modul zobnika ###

```python
# vhodni podatki za enačbe modula (skupni) #
psi_d = 0.6
beta = 20 * np.pi/180
sig_Flim1 = 430 #MPa  # Material: 16MnCr5
sig_Hlim1 = 1470 #MPa
K_gamma = 1.2

### Prva zobniška stopnja ###
# vhodni podatki za enačbe modula pri prvi stopnji#
M_t1 = P_b/(2*np.pi*n_m/60)*1000 # N mm - glede na moment motorja
u = z_21/z_11
n_p = n_p1
# Moduli določeni glede na korensko trajno dinamično trdnost F-lim in glede na bočno trajno dinamično trdnost H-lim
m_nH = 95*np.cos(beta)/z_11 * np.cbrt( M_t1*K_gamma/(n_p*psi_d*sig_Hlim1**2)* (u+1)/u )
m_nF = 1.85*np.cbrt( M_t1*K_gamma*np.cos(beta)**2/(n_p*z_11**2*psi_d*sig_Flim1) )
m_nH = np.around(m_nH, 3)
m_nF = np.around(m_nF, 3)
# Izbira standardnega modula glede na rezultat
m_n1 = 1.25
print("PRVA ZOBNIŠKA STOPNJA:")
print("Izberemo višjo vrednost in zaokrožimo na standardni modul.")
print(f"Glede na bočno trdnost: m_nH = {m_nH} mm, Glede na korensko trdnost: m_nF = {m_nF} mm.")
print(f"Izberemo modul m_n = {m_n1} mm.")
print()

### Druga zobniška stopnja ###
# vhodni podatki za enačbe modula pri drugi stopnji #
M_t2 = np.abs(M_ts1) # N mm - glede na moment motorja
beta = 20 * np.pi/180
u = 1
n_p = n_p2
# Moduli določeni glede na korensko trajno dinamično trdnost F-lim in glede na bočno trajno dinamično trdnost H-lim
m_nH = 95*np.cos(beta)/z_12 * np.cbrt( M_t2*K_gamma/(n_p*psi_d*sig_Hlim1**2)* (u+1)/u )
m_nF = 1.85*np.cbrt( M_t2*K_gamma*np.cos(beta)**2/(n_p*z_12**2*psi_d*sig_Flim1) )
m_nH = np.around(m_nH, 3)
m_nF = np.around(m_nF, 3)
# Izbira standardnega modula glede na rezultat
m_n2 = 1.5
print("DRUGA ZOBNIŠKA STOPNJA:")
print("Izberemo višjo vrednost in zaokrožimo na standardni modul.")
print(f"Glede na bočno trdnost: m_nH = {m_nH} mm, Glede na korensko trdnost: m_nF = {m_nF} mm.")
print(f"Izberemo modul m_n = {m_n2} mm.")
print()

### Tretja zobniška stopnja ###
# vhodni podatki za enačbe modula pri tretji stopnji #
M_t3 = np.abs(M_ts2) # N mm - glede na moment motorja
beta = 20 * np.pi/180
u = 1
n_p = n_p3
# Moduli določeni glede na korensko trajno dinamično trdnost F-lim in glede na bočno trajno dinamično trdnost H-lim
m_nH = 95*np.cos(beta)/z_13 * np.cbrt( M_t3*K_gamma/(n_p*psi_d*sig_Hlim1**2)* (u+1)/u )
m_nF = 1.85*np.cbrt( M_t3*K_gamma*np.cos(beta)**2/(n_p*z_13**2*psi_d*sig_Flim1) )
m_nH = np.around(m_nH, 3)
m_nF = np.around(m_nF, 3)
# Izbira standardnega modula glede na rezultat
m_n3 = 2.0
print("TRETJA ZOBNIŠKA STOPNJA:")
print("Izberemo višjo vrednost in zaokrožimo na standardni modul.")
print(f"Glede na bočno trdnost: m_nH = {m_nH} mm, Glede na korensko trdnost: m_nF = {m_nF} mm.")
print(f"Izberemo modul m_n = {m_n3} mm.")
```
